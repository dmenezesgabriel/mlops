# NYC Taxi Demand Forecasting: MLOps Lifecycle

Welcome to the MLOps Lab! This project is a showcase of a production-quality, end-to-end Machine Learning Operations (MLOps) lifecycle.

We implement the principles of **Continuous Delivery for Machine Learning (CD4ML)** to build a reproducible, automated, and audited system using industry-standard tools:
- **Feast** as our Feature Store.
- **MLflow** as our Model Registry & Experiment Tracking store.
- **DuckDB** and **SQLite** as our local data stores.
- **FastAPI** for real-time online inference serving.

1. **[[problem-framing|Problem Framing]]**: Framing the business objective and prediction task.

4. **[[data-preprocessing|Data Preprocessing]]**: Cleaning transaction logs and grouping into hourly demand.
5. **[[feature-engineering|Feature Engineering]]**: Managing features with Feast entities and views.
6. **[[model-training|Train]]**: Fitting a Ridge Regressor on historical features.
7. **[[hyperparameter-tuning|Tune]]**: Optimizing parameters via MLflow sweeps.
8. **[[model-evaluation|Evaluate]]**: Enforcing quality gates on the candidate models.
9. **[[model-deployment|Deploy]]**: Serving models in real-time or via batch scoring.
10. **[[model-monitoring|Monitoring]]**: Detecting drift and observing metrics.

## How to run the project locally
Each stage of the MLOps lifecycle is executable via `make` targets. To run the entire pipeline end-to-end:
```bash
make collect
make preprocess
make features
make tune
make train
make evaluate
make deploy
make monitor
```
